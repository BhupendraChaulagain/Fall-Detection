```python
# Scaling data to a  range of [-10, 10] and [-20,20] for proper visualization
def scale(data, min_range, max_range):
    data_min = np.min(data)
    data_max = np.max(data)
    return ((data - data_min) / (data_max - data_min)) * (max_range - min_range) + min_range

```

```python
# Scaling gyroscope data
fall1['Gx'] = scale(fall1['Gx'], -20, 20)
fall1['Gy'] = scale(fall1['Gy'], -20, 20)
fall1['Gz'] = scale(fall1['Gz'], -20, 20)

```

```python
# Scaling accelerometer data
fall1['Ax'] = scale(fall1['Ax'], -10, 10)
fall1['Ay'] = scale(fall1['Ay'], -10, 10)
fall1['Az'] = scale(fall1['Az'], -10, 10)
```

```python
# Plotting for Accelerometer and Gyroscope Data
plt.figure(figsize=(12, 6))

plt.plot(range(len(fall1['Gx'])), fall1['Gx'], label='Gx', color='red', linewidth=1.5, alpha=0.7)
plt.plot(range(len(fall1['Gy'])), fall1['Gy'], label='Gy', color='green', linewidth=1.5, alpha=0.7)
plt.plot(range(len(fall1['Gz'])), fall1['Gz'], label='Gz', color='black', linewidth=1.5, alpha=0.7)
plt.plot(range(len(fall1['Ax'])), fall1['Ax'], label='Ax', color='blue', linewidth=1.5, alpha=0.7)
plt.plot(range(len(fall1['Ay'])), fall1['Ay'], label='Ay', color='purple', linewidth=1.5, alpha=0.7)
plt.plot(range(len(fall1['Az'])), fall1['Az'], label='Az', color='orange', linewidth=1.5, alpha=0.7)

# Highlighting regions for falls
for i in range(len(fall1['Label'])):
    if fall1['Label'][i] == 1:  # Occurance of Fall event
        plt.axvspan(i - 0.5, i + 0.5, color='orange', alpha=0.3)  # Highlighting the fall

plt.title('Scaled Accelerometer and Gyroscope Data with Fall Indications')
plt.xlabel('Index')
plt.ylabel('Scaled Value')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
